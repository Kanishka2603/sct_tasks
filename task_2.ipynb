```python
import pandas as pd

file_path = "/content/drive/MyDrive/Titanic-Dataset  task_2.csv"
df = pd.read_csv(file_path)

# Display basic info
df.info()
df.head()

df.isnull().sum()
df['Age'].fillna(df['Age'].median(), inplace=True)  # Fill missing ages with median
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)  # Fill missing embarkation port with mode
df.drop(columns=['Cabin'], inplace=True)  # Drop 'Cabin' due to many missing values
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=df["Sex"], y=df["Survived"])
plt.title("Survival Rate by Gender")
plt.show()

sns.barplot(x=df["Pclass"], y=df["Survived"])
plt.title("Survival Rate by Passenger Class")
plt.show()

sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()




```
