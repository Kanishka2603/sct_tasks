```python
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
  # Update this path if needed
df = pd.read_csv("/content/drive/MyDrive/world_population task_1.csv")
```

```python
# Select top 10 most populous countries in 2022
top_10 = df.nlargest(10, '2022 Population')

# Create bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10["Country/Territory"], top_10["2022 Population"], color='royalblue')

# Labels and title
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.title("Top 10 Most Populous Countries in 2022", fontsize=14)
plt.xticks(rotation=45)

# Show plot
plt.show()
```
